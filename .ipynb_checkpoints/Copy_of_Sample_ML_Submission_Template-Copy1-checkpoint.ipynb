### 4. Feature Manipulation & Selection

#### 1. Feature Manipulation

````python
# Manipulate Features to minimize feature correlation and create new features
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(x):
    vif=pd.DataFrame()
    vif['variables']=x.columns
    vif['VIF']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return vif
````

#### 2. Feature Selection

````python
# Select your features wisely to avoid overfitting
calculate_vif(df[[i for i in df.describe().columns if i not in ['Rented Bike Count']]])
````

````python
calculate_vif(df[[i for i in df.describe().columns if i not in ['Rented Bike Count',\
                                                                'Dew point temperature(°C)','Humidity(%)','Visibility (10m)']]])
````

````python
features =list(set(df.describe().columns)-set( ['Rented Bike Count','Dew point temperature(°C)',\
                                      'Humidity(%)','Visibility (10m)']))
features
                                
````

````python
df.iloc[0]
````

````python

# Encoding
df['Holiday']=df['Holiday'].apply(lambda x: 1 if x=='Holiday' else 0)
df['Functioning Day']=df['Functioning Day'].apply(lambda x: 1 if x=='Yes' else 0)
df['day_weekend']=df['day_weekend'].apply(lambda x: 1 if x==['Sunday','Saturday'] else 0)

df=pd.get_dummies(df,columns=['Seasons','Hour'])
````

````python
X=df.iloc[:,1:]
X
````

````python
y=df.iloc[:,0]
y
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
scaler=StandardScaler()

X_train=scaler.fit_transform(X_train)
X_train
````

````python
X_test=scaler.transform(X_test)
````

````python
def calc_scores(true,pred):
    MAE=mean_absolute_error(true,pred)
    print(f'MAE is {MAE}')
    
    MSE=mean_squared_error(true,pred)
    print(f'MSE is {MSE}')
    
    RMSE=np.sqrt(MSE)
    print(f'RMSE is {RMSE}')
    
    R2=r2_score(true,pred)
    print(f'R2 is {R2}')
    
````

````python
score_df=pd.DataFrame({'Model':[],'Mean AE':[],'MSE':[],'RMSE':[],'R2 Score':[]})
def store_scores(model,mae,mse,rmse,r2):
    scores={'Model':model,'Mean AE':mae,'MSE':mse,'RMSE':rmse,'R2 Score':r2}
    global score_df
    score_df=score_df.append(scores,ignore_index=True)
    return score_df
````

````python
def reg_scatter(true,pred):
    plt.figure(figsize=(10,8))
    
    sns.regplot(x=true,y=pred,scatter_kws={'color':'magenta'},line_kws={'color':'black'})
    plt.legend(['Actual','Predicted'])
    plt.show()
````

````python
def feature_imp(model):
    model1=[model]
    best_rf=model.best_estimator_
    feature_importances=best_rf.feature_importances_
    
    feature_importance_df=pd.DataFrame({'Feature':X.columns,'Importance':feature_importances})
    feature_importance_df=feature_importance_df.sort_values('Importance',ascending=False)
    
    sns.set_style('whitegrid')
    plt.figure(figsize=(15,8))
    sns.barplot(x='Importance',y='Feature',data=feature_importance_df,order=feature_importance_df['feature'])
    if model1==[model]:
        plt.title("Hi1")
    else:
        plt.title("Hi2")
    plt.xlabel("Importance")
    plt.ylabel('Features')
    plt.show()
````

````python
lin_reg=LinearRegression()
reg=lin_reg.fit(X_train,y_train)

y_train_pred=reg.predict(X_train)
y_test_pred=reg.predict(X_test)
print('Coeff is',reg.coef_)
print("Intercept is",reg.intercept_)
````

````python
print('LR model train score is',reg.score(X_train,y_train))
print('LR model test score is',reg.score(X_test,y_test))
````

````python
calc_scores(y_train,y_train_pred)
````

````python
reg_scatter(y_test,y_test_pred)
# scatter points are plotte in (num of features in X+1)dimensions
````
